```python
def remove_duplicate_columns(df):
    """
    Remove duplicate columns from DataFrame as specified.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input DataFrame
    
    Returns:
    --------
    pandas.DataFrame
        DataFrame with specified duplicate columns removed
    """
    # Create a copy to avoid modifying original
    df_clean = df.copy()
    
    # List of all columns to remove
    columns_to_remove = [
        # Auto-generated indices
        'Unnamed: 0',
        'index',
        
        # Duplicate stats (keeping _x versions, removing _y versions)
        'pass_cmp_y',
        'rush_att_y',
        'rush_yds_y', 
        'rush_td_y',
        'targets_y',
        'rec_y',
        'rec_yds_y',
        'rec_td_y',
        'def_int_y',
        'sacks_y',
        'tackles_combined_y',
        
        # Duplicate without suffix
        ' pass_att',  # with leading space
        
        # Keep elevation, remove stadium (or vice versa - choose one)
        'team_home_stadium',  # Remove this, keep team_home_elevation
    ]
    
    # Remove the columns
    df_clean = df_clean.drop(columns=columns_to_remove, errors='ignore')
    
    return df_clean
```

```python
clean_left_merged_data = remove_duplicate_columns(left_merged_data)
clean_right_merged_data = remove_duplicate_columns(right_merged_data)
clean_inner_merged_data = remove_duplicate_columns(inner_merged_data)
clean_outer_merged_data = remove_duplicate_columns(outer_merged_data)

print(len(right_merged_data.columns))
print(len(clean_right_merged_data.columns))
```
