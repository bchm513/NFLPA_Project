```python
# Add travel direction and magnitude
def add_travel_features(df, tz_diff_col='game_tz_difference'):
    """
    Add comprehensive travel features based on timezone difference
    """
    df = df.copy()
    
    # 1. Basic travel direction
    df['travel_direction'] = np.select(
        [
            df[tz_diff_col] == 0,
            df[tz_diff_col] > 0,
            df[tz_diff_col] < 0
        ],
        ['home', 'east', 'west'],
        default='unknown'
    )
    
    # 2. Absolute timezone difference (magnitude)
    df['tz_diff_magnitude'] = df[tz_diff_col].abs()
    
    # 3. Travel category based on magnitude
    conditions = [
        df[tz_diff_col] == 0,
        df['tz_diff_magnitude'] == 1,
        df['tz_diff_magnitude'] == 2,
        df['tz_diff_magnitude'] == 3,
        df['tz_diff_magnitude'] >= 4
    ]
    
    categories = ['home', '1_hr', '2_hr', '3_hr', '4+_hr']
    df['travel_magnitude'] = np.select(conditions, categories, default='unknown')
    
    # 4. Is long travel? (3+ hours difference)
    df['is_long_travel'] = df['tz_diff_magnitude'] >= 3
    
    # 5. Detailed travel description
    def get_travel_description(row):
        tz_diff = row[tz_diff_col]
        if pd.isna(tz_diff):
            return 'unknown'
        if tz_diff == 0:
            return 'home_game'
        direction = 'east' if tz_diff > 0 else 'west'
        hours = abs(int(tz_diff))
        return f'travel_{direction}_{hours}_hours'
    
    df['travel_description'] = df.apply(get_travel_description, axis=1)
    
    return df

# Apply the function
combined_data = add_travel_features(combined_data, 'game_tz_difference')

# Analyze the results
print("Travel analysis:")
print("=" * 50)

# Direction breakdown
print("\n1. Travel Direction:")
print(combined_data['travel_direction'].value_counts().sort_index())

# Magnitude breakdown
print("\n2. Travel Magnitude:")
print(combined_data['travel_magnitude'].value_counts().sort_index())

# Long travel analysis
print("\n3. Long Travel Analysis:")
long_travel_pct = combined_data['is_long_travel'].mean() * 100
print(f"Games with long travel (3+ hours): {long_travel_pct:.1f}%")

# Most common travel scenarios
print("\n4. Most Common Travel Scenarios:")
common_scenarios = combined_data['travel_description'].value_counts().head(10)
for scenario, count in common_scenarios.items():
    pct = (count / len(combined_data)) * 100
    print(f"  {scenario}: {count} games ({pct:.1f}%)")

# Sample output
print("\n5. Sample Data:")
sample = combined_data[['player_team', 'game_tz_difference', 'travel_direction', 
                       'travel_magnitude', 'travel_description']].head(10)
print(sample)
```

```python
# %pip install geopy

from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time

# First, create a comprehensive stadium location dictionary
# Coordinates are (latitude, longitude)
stadium_locations = {
    # Eastern Time (ET) - GMT-5
    'Raymond James Stadium': (27.9759, -82.5033),  # Tampa, FL
    'Mercedes-Benz Stadium': (33.7555, -84.4007),  # Atlanta, GA
    'Highmark Stadium': (42.7737, -78.7870),  # Buffalo, NY
    'Bank of America Stadium': (35.2258, -80.8528),  # Charlotte, NC
    'Paul Brown Stadium': (39.0955, -84.5160),  # Cincinnati, OH
    'Lucas Oil Stadium': (39.7601, -86.1639),  # Indianapolis, IN
    'Nissan Stadium': (36.1665, -86.7713),  # Nashville, TN
    'Ford Field': (42.3400, -83.0456),  # Detroit, MI
    'NRG Stadium': (29.6847, -95.4108),  # Houston, TX (CT)
    'FedExField': (38.9078, -76.8644),  # Landover, MD
    'MetLife Stadium': (40.8135, -74.0745),  # East Rutherford, NJ
    'TIAA Bank Stadium': (30.3239, -81.6373),  # Jacksonville, FL
    'Gillette Stadium': (42.0909, -71.2643),  # Foxborough, MA
    'Hard Rock Stadium': (25.9581, -80.2389),  # Miami Gardens, FL
    'FirstEnergy Stadium': (41.5061, -81.6995),  # Cleveland, OH
    'Lincoln Financial Field': (39.9008, -75.1675),  # Philadelphia, PA
    'Heinz Field': (40.4468, -80.0158),  # Pittsburgh, PA
    'M&T Bank Stadium': (39.2781, -76.6228),  # Baltimore, MD
    'Paycor Stadium': (39.0955, -84.5160),  # Cincinnati, OH (same as Paul Brown)
    'Acrisure Stadium': (40.4468, -80.0158),  # Pittsburgh, PA (same as Heinz)
    'Cleveland Browns Stadium': (41.5061, -81.6995),  # Cleveland, OH (same as FirstEnergy)
    'EverBank Stadium': (30.3239, -81.6373),  # Jacksonville, FL (same as TIAA Bank)
    'Northwest Stadium': (38.9078, -76.8644),  # Maryland (same as FedExField approx)
    'Georgia Dome': (33.7574, -84.4008),  # Atlanta, GA (approx)
    'New Era Field': (42.7737, -78.7870),  # Buffalo, NY (same as Highmark)
    'EverBank Field': (30.3239, -81.6373),  # Jacksonville, FL (same as TIAA Bank)
    'Bills Stadium': (42.7737, -78.7870),  # Buffalo, NY (same as Highmark)
    
    # Central Time (CT) - GMT-6
    'Lambeau Field': (44.5013, -88.0622),  # Green Bay, WI
    'GEHA Field at Arrowhead Stadium': (39.0489, -94.4839),  # Kansas City, MO
    'AT&T Stadium': (32.7473, -97.0945),  # Arlington, TX
    'Soldier Field': (41.8623, -87.6167),  # Chicago, IL
    'Huntington Bank Field': (41.8623, -87.6167),  # Chicago, IL (same as Soldier)
    'U.S. Bank Stadium': (44.9737, -93.2580),  # Minneapolis, MN
    'Caesars Superdome': (29.9508, -90.0811),  # New Orleans, LA
    'Arrowhead Stadium': (39.0489, -94.4839),  # Kansas City, MO (same as GEHA)
    'Mercedes-Benz Superdome': (29.9508, -90.0811),  # New Orleans (same as Caesars)
    
    # Mountain Time (MT) - GMT-7
    'State Farm Stadium': (33.5275, -112.2626),  # Glendale, AZ
    'Empower Field at Mile High': (39.7439, -105.0201),  # Denver, CO
    'Sports Authority Field at Mile High': (39.7439, -105.0201),  # Denver, CO
    'University of Phoenix Stadium': (33.5275, -112.2626),  # Glendale, AZ (same as State Farm)
    
    # Pacific Time (PT) - GMT-8
    'SoFi Stadium': (33.9535, -118.3392),  # Inglewood, CA
    'Lumen Field': (47.5952, -122.3316),  # Seattle, WA
    "Levi's Stadium": (37.4030, -121.9700),  # Santa Clara, CA
    'Allegiant Stadium': (36.0908, -115.1834),  # Las Vegas, NV
    'CenturyLink Field': (47.5952, -122.3316),  # Seattle, WA (same as Lumen)
    'Los Angeles Memorial Coliseum': (34.0141, -118.2879),  # Los Angeles, CA
    'Oakland-Alameda County Coliseum': (37.7516, -122.2005),  # Oakland, CA
    'Qualcomm Stadium': (32.7831, -117.1195),  # San Diego, CA
    'StubHub Center': (33.8644, -118.2611),  # Carson, CA
    'Ring Central Coliseum': (37.7516, -122.2005),  # Oakland, CA (same as Oakland-Alameda)
    
    # Non-US stadiums
    'Tottenham Stadium': (51.6043, -0.0664),  # London, UK
    'Wembley Stadium': (51.5560, -0.2795),  # London, UK
    'Allianz Arena': (48.2188, 11.6247),  # Munich, Germany
    'Azteca Stadium': (19.3030, -99.1504),  # Mexico City, Mexico
    'Deutsche Bank Park': (50.0686, 8.6454),  # Frankfurt, Germany
    'Neo Quimica Arena': (-23.5450, -46.4744),  # São Paulo, Brazil
    'Croke Park': (53.3608, -6.2515),  # Dublin, Ireland
    'Olympiastadion': (52.5147, 13.2395),  # Berlin, Germany
    'Estadio Santiago Bernabéu': (40.4531, -3.6883),  # Madrid, Spain
    'Twickenham Stadium': (51.4560, -0.3412),  # London, UK
}

# Now create a team home stadium mapping
team_home_stadiums = {
    'Atlanta Falcons': 'Mercedes-Benz Stadium',
    'Baltimore Ravens': 'M&T Bank Stadium',
    'Buffalo Bills': 'Highmark Stadium',
    'Carolina Panthers': 'Bank of America Stadium',
    'Cincinnati Bengals': 'Paycor Stadium',
    'Cleveland Browns': 'Cleveland Browns Stadium',
    'Dallas Cowboys': 'AT&T Stadium',
    'Denver Broncos': 'Empower Field at Mile High',
    'Detroit Lions': 'Ford Field',
    'Green Bay Packers': 'Lambeau Field',
    'Houston Texans': 'NRG Stadium',
    'Indianapolis Colts': 'Lucas Oil Stadium',
    'Jacksonville Jaguars': 'EverBank Stadium',
    'Kansas City Chiefs': 'GEHA Field at Arrowhead Stadium',
    'Las Vegas Raiders': 'Allegiant Stadium',
    'Los Angeles Chargers': 'SoFi Stadium',
    'Los Angeles Rams': 'SoFi Stadium',
    'Miami Dolphins': 'Hard Rock Stadium',
    'Minnesota Vikings': 'U.S. Bank Stadium',
    'New England Patriots': 'Gillette Stadium',
    'New Orleans Saints': 'Caesars Superdome',
    'New York Giants': 'MetLife Stadium',
    'New York Jets': 'MetLife Stadium',
    'Oakland Raiders': 'Oakland-Alameda County Coliseum',
    'Philadelphia Eagles': 'Lincoln Financial Field',
    'Pittsburgh Steelers': 'Acrisure Stadium',
    'San Diego Chargers': 'Qualcomm Stadium',
    'San Francisco 49ers': "Levi's Stadium",
    'Seattle Seahawks': 'Lumen Field',
    'Tampa Bay Buccaneers': 'Raymond James Stadium',
    'Tennessee Titans': 'Nissan Stadium',
    'Washington Football Team': 'FedExField',
    'Washington Redskins': 'FedExField',
    'Arizona Cardinals': 'State Farm Stadium',
    'Chicago Bears': 'Soldier Field',
}

# Function to calculate distance between two stadiums
def calculate_distance(stadium1, stadium2):
    """Calculate distance in miles between two stadiums"""
    if pd.isna(stadium1) or pd.isna(stadium2):
        return np.nan
    
    if stadium1 not in stadium_locations or stadium2 not in stadium_locations:
        return np.nan
    
    try:
        coords1 = stadium_locations[stadium1]
        coords2 = stadium_locations[stadium2]
        # Calculate distance in miles
        distance_miles = geodesic(coords1, coords2).miles
        return distance_miles
    except:
        return np.nan

# Add columns to your DataFrame
def add_distance_features(df):
    """
    Add distance traveled features to DataFrame
    """
    df = df.copy()
    
    # 1. Get team's home stadium
    df['team_home_stadium'] = df['player_team'].map(team_home_stadiums)
    
    # 2. Calculate distance traveled (from home stadium to game stadium)
    df['distance_traveled_miles'] = df.apply(
        lambda row: calculate_distance(row['team_home_stadium'], row['stadium']),
        axis=1
    )
    
    # 3. Add kilometers version
    df['distance_traveled_km'] = df['distance_traveled_miles'] * 1.60934
    
    # 4. Categorize distance
    bins = [0, 100, 300, 500, 1000, 2000, 5000, float('inf')]
    labels = ['<100 mi', '100-300 mi', '300-500 mi', '500-1000 mi', 
              '1000-2000 mi', '2000-5000 mi', '>5000 mi']
    
    df['distance_category'] = pd.cut(
        df['distance_traveled_miles'], 
        bins=bins, 
        labels=labels,
        right=False
    )
    
    # 5. Is long distance travel? (500+ miles)
    df['is_long_distance'] = df['distance_traveled_miles'] >= 500
    
    # 6. Is cross-country travel? (2000+ miles)
    df['is_cross_country'] = df['distance_traveled_miles'] >= 2000
    
    # 7. Is international travel? (outside US stadiums)
    df['is_international_travel'] = df['stadium'].isin(non_us_stadiums)
    
    # 8. Flight time estimate (assuming 500 mph cruising speed + 2 hours for airport time)
    df['estimated_travel_hours'] = df.apply(
        lambda row: (row['distance_traveled_miles'] / 500) + 2 
        if not pd.isna(row['distance_traveled_miles']) else np.nan,
        axis=1
    )
    
    # 9. Travel fatigue index (combines distance and timezone difference)
    if 'game_tz_difference' in df.columns:
        df['travel_fatigue_index'] = (
            (df['distance_traveled_miles'] / 1000) +  # Normalize distance
            df['game_tz_difference'].abs() * 0.5      # Add timezone penalty
        )
    
    return df
```
