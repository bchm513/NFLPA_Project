```python
import pandas as pd

injury_data = pd.read_csv('player_injury_data.csv')
stats_data = pd.read_csv('player_snap_data_tz_miles.csv')
```

```python
injury_data["date"] = (
    injury_data["date"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.lower()
)

stats_data["date"] = (
    stats_data["date"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.lower()
)

injury_data["player_team"] = injury_data["player_team"].str.split().str[-1]

stats_data["player_team"] = stats_data["player_team"].str.split().str[-1]
```

```python
# Create a mapping dictionary
month_mapping = {
    'january': 'jan',
    'february': 'feb', 
    'march': 'mar',
    'april': 'apr',
    'may': 'may',
    'june': 'jun',
    'july': 'jul',
    'august': 'aug',
    'september': 'sep',
    'october': 'oct',
    'november': 'nov',
    'december': 'dec'
}

# Apply to your dataframe
injury_data['date'] = injury_data['date'].copy()

for full, abbr in month_mapping.items():
    injury_data['date'] = injury_data['date'].str.replace(full, abbr, case=False, regex=False)

print(injury_data[['date']].head())
```

```python
merged_data = pd.merge(
    injury_data, 
    stats_data, 
    on=['player_name', 'date', 'player_team'],
    how='left' 
)

merged_data
```
